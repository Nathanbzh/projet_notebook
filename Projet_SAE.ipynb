```python
# Filtrer les salaires pour un pays donné (ex : US)
data_us = data[data['company_location'] == 'US']
print(data_us.head())
```

```python
# Top 5 des entreprises qui paient le plus
top_5_pays_salaire = data.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False).head(5)
print(top_5_pays_salaire)
```

```python
# Filtrer les données
# Rôles avec un salaire moyen supérieur à 100,000
high_salary = data[data['salary_in_usd'] > 100000][['job_title', 'salary_in_usd']]
print(high_salary.head())

# Pays avec la plus forte proportion de télétravail (remote_ratio = 100)
remote_country = data.groupby('company_location')['remote_ratio'].mean().sort_values(ascending=False).head(5)
print(remote_country)
```

```python
# Requêtes conditionnelles
# Salaire moyen des employés à distance (100%)
avg_salary_remote = data[data['remote_ratio'] == 100]['salary_in_usd'].mean()
print(avg_salary_remote)

# Comparer salaires moyens pour deux catégories
mean_se = data[data['experience_level'] == 'SE']['salary_in_usd'].mean()
mean_mi = data[data['experience_level'] == 'MI']['salary_in_usd'].mean()
print(f"Moyenne Senior : {mean_se:.2f} vs Moyenne Mid-level : {mean_mi:.2f}")
```

```python
# Ajouter une colonne pour les fourchettes de salaires
bins = [0, 50000, 100000, 150000, 200000, np.inf]
labels = ['<50k', '50k-100k', '100k-150k', '150k-200k', '200k+']

data['salary_range'] = pd.cut(data['salary_in_usd'], bins=bins, labels=labels)
print(data[['salary_in_usd', 'salary_range']].head())
```

```python
# Analyse des corrélations
# Relation entre télétravail et salaire
correlation = data[['remote_ratio', 'salary_in_usd']].corr()
print("Matrice de corrélation :\n", correlation)
```

```python
# Salaire moyen par taille d'entreprise
avg_salary_size = data.groupby('company_size')['salary_in_usd'].mean()
print("\nSalaire moyen par taille d'entreprise :\n", avg_salary_size)
```

```python
# Tableaux croisés
# Moyenne des salaires par niveau d'expérience et télétravail
pivot_table = data.pivot_table(
    index='experience_level', 
    columns='remote_ratio', 
    values='salary_in_usd', 
    aggfunc='mean'
)
print("Tableau croisé des salaires :")
print(pivot_table)
```
